## 3. Quick look at the data (2-3 real numbers)

The starter dataset gives enough real signals to justify this lane. I will check how many pages exist, how many are declining, how many declining pages still have search demand, and how many visible pages have low CTR. These numbers help show why a refresh-prioritization queue can be useful.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import os
import subprocess
import pandas as pd
import numpy as np

DATA_PATH = "data/raw/content_refresh_anonymized.csv"

# If running in Colab and the data file is not available, clone my repo.
if not os.path.exists(DATA_PATH):
    REPO_URL = "https://github.com/Moezulhaq24/FlyRank-Internship-ML"
    REPO_DIR = "FlyRank-Internship-ML"

    if not os.path.isdir(REPO_DIR):
        subprocess.run(["git", "clone", "--depth", "1", REPO_URL, REPO_DIR], check=True)

    os.chdir(REPO_DIR)

df = pd.read_csv(DATA_PATH)

declining = df["trend_direction"].str.lower().eq("down")
visible_pages = df["impressions_90d"] >= 500
declining_with_demand = declining & (df["impressions_90d"] >= 100)

low_ctr_visible = (
    (df["impressions_90d"] >= 500)
    & (df["avg_position"] > 0)
    & (df["avg_position"] <= 20)
    & (df["ctr"] < 0.5)
)

low_engagement_visible = (
    (df["sessions_90d"] >= 30)
    & (
        ((df["engagement_rate"] > 0) & (df["engagement_rate"] < 30))
        | ((df["scroll_rate"] > 0) & (df["scroll_rate"] < 30))
    )
)

search_volume_corr = df["search_volume"].corr(df["impressions_90d"])

print("Total pages:", len(df))
print("Total columns:", df.shape[1])
print("Total pseudonymized clients:", df["client_id"].nunique())

print("\nDeclining pages:", declining.sum())
print("Declining rate:", round(declining.mean(), 3))

print("\nVisible pages with impressions_90d >= 500:", visible_pages.sum())
print("Declining pages with impressions_90d >= 100:", declining_with_demand.sum())
print("Visible pages with low CTR:", low_ctr_visible.sum())
print("Visible pages with weak engagement:", low_engagement_visible.sum())

print("\nCorrelation between search_volume and impressions_90d:", round(search_volume_corr, 4))
```

##Result Explanation

These numbers support my lane choice. The starter dataset contains 30,000 pages across 32 pseudonymized clients. Around 54.2% of pages are marked as declining in the starter label, which means there is enough decline signal to study.

There are also many pages where action could matter: 13,152 declining pages still have at least 100 impressions, and 9,759 visible pages have low CTR. This suggests that content review should not only look at whether a page is old or declining. It should also consider demand, CTR, position, engagement, and freshness together.

The correlation between search volume and actual page impressions is almost zero, so keyword volume alone is not enough to decide which pages deserve review first. This supports using a data-driven ranking approach instead of one simple rule.
